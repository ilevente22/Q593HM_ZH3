```python
# Csomagimportok

import pandas as pd
```

```python
# Adatok kiíratása

data = pd.read_csv("avocado.csv").dropna()
```

```python
# Régiók kiírása

unique_values = data['region'].unique()
print(unique_values)
```

```python
#Hány régióból származnak az adatok
print("Hányféle régióból származnak adatok az adathalmazban?", len(unique_values))
```

```python
# Egy régióra átlagban hány avokádó jut.
avocado_mean = data.groupby('region')['Total Volume'].agg(['mean'])
print(f"Napi avokádóeladás átlaga régiónként: {avocado_mean}")

# Egy régióra jutó avokádók mediánja.
avocado_median = data.groupby('region')['Total Volume'].agg(['median'])
print(f"Napi avokádóeladás mediánja régiónként: {avocado_median}.")

# Egy régióra jutó avokádók szórása.
avocado_std = data.groupby('region')['Total Volume'].agg(['std'])
print(f"Napi avokádóeladás szórása régiónként {avocado_std}")
```

```python
# Az összes eladott avokádó mennyiségének kiszámítása évenként és régiónként
total_sold = data.groupby(['year', 'region'])['Total Volume'].sum().reset_index(name='Total_Sold')

# A hagyományos avokádók eladásainak kiszámítása
conventional_sold = data[data['type'] == 'conventional'].groupby(['year', 'region'])['Total Volume'].sum().reset_index(name='Conventional_Sold')

# Az eredmények összevonása egy DataFrame-be
final_df = total_sold.merge(conventional_sold, on=['year', 'region'])

# A hagyományos avokádó eladások százalékos arányának kiszámítása
final_df['Conventional_Percent'] = (final_df['Conventional_Sold'] / final_df['Total_Sold'])

# A bio avokádó eladások százalékos arányának kiszámítása
final_df['Organic_Percent'] = 1 - final_df['Conventional_Percent']

# Az átlag, medián és szórás kiszámítása a százalékos arányokra vonatkozóan
aggregated_data = final_df.groupby(['year']).agg({'Conventional_Percent': ['mean', 'median', 'std'], 'Organic_Percent': ['mean', 'median', 'std']}).reset_index()

# Az eredmények átalakítása könnyen olvasható formátumba
aggregated_data.columns = ['Év', 'Hagyományos_Átlag', 'Hagyományos_Medián', 'Hagyományos_Szórás', 'Bio_Átlag', 'Bio_Medián', 'Bio_Szórás']

print(aggregated_data)
```
